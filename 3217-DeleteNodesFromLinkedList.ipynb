```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def modifiedList(self, nums: List[int], head: Optional[ListNode]) -> Optional[ListNode]:
        if not head:
            return head

        nums_set = set(nums)       #silinecek degerler
        dummy = ListNode(0, head)  #head'in onune degeri 0 olan node ekle, bu sayede head'i silmek kolay olur
        prev = dummy
        curr = head

        while curr:
            if curr.val in nums_set:
                prev.next = curr.next
                curr = curr.next
            else:
                # pointer'lari ilerlet
                prev = curr
                curr = curr.next

        return dummy.next
```
